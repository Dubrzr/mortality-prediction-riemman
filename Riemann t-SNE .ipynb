```python
from copy import copy
import os
import json
import numpy
from pyriemann.utils.distance import pairwise_distance, distance_riemann
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
import matplotlib as mpl
import matplotlib.colors as colors
import pandas
import itertools
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report


# Colors
lengths = [0, 60*15, 60*30, 60*45, 60*60, 60*60*2, 60*60*4, 60*60*12, 60*60*24, 60*60*24*2, 60*60*24*3,60*60*24*7, 60*60*24*15, 60*60*24*30]
bounds = numpy.array(lengths)
palette = copy(plt.cm.rainbow)
palette.set_over('red', 1.0)
palette.set_under('black', 0)
palette.set_bad('black', 0)
norm = colors.BoundaryNorm(boundaries=bounds, ncolors=palette.N, clip=False)
    
def beautify_length(length, fs=125, short=False):
    length_sec = length/fs
    if short:
        if length_sec > 60*60*24*365:
            return "{}y".format(int(length_sec/(60*60*24)))
        if length_sec > 60*60*24*30:
            return "{}mo".format(int(length_sec/(60*60*24)))
        if length_sec > 60*60*24*7:
            return "{}w".format(int(length_sec/(60*60*24)))
        if length_sec > 60*60*24:
            return "{}d".format(int(length_sec/(60*60*24)))
        if length_sec > 60*60:
            return "{}h".format(int(length_sec/(60*60)))
        if length_sec > 60:
            return "{}mi".format(int(length_sec/(60)))
        return "{}s".format(int(length_sec))
    else:
        if length_sec > 60*60*24*365:
            return "{} years".format(int(length_sec/(60*60*24)))
        if length_sec > 60*60*24*30:
            return "{} months".format(int(length_sec/(60*60*24)))
        if length_sec > 60*60*24*7:
            return "{} weeks".format(int(length_sec/(60*60*24)))
        if length_sec > 60*60*24:
            return "{} days".format(int(length_sec/(60*60*24)))
        if length_sec > 60*60:
            return "{} hours".format(int(length_sec/(60*60)))
        if length_sec > 60:
            return "{} minutes".format(int(length_sec/(60)))
        return "{} seconds".format(int(length_sec))

neighbors_classes = [(0, 60*30), (60*30, 60*60*2), (60*60*2, 60*60*4), (60*60*4, 60*60*12), (60*60*12, 60*60*24*2), (60*60*24*2, 60*60*24*7), (60*60*24*7, 60*60*24*15), (60*60*24*15, 60*60*24*30),  (60*60*24*30, 60*60*24*6000)]
neighbors_classes_names = [beautify_length(a*125, short=True) + ' to ' + beautify_length(b*125, short=True) for a,b in neighbors_classes]
# neighbors_classes_names = [str(a) + ' to ' + str(b) for a,b in neighbors_classes]


def my_dist(a,b):
    a = numpy.reshape(a, (int(numpy.sqrt(a.shape[0])),int(numpy.sqrt(a.shape[0]))))
    b = numpy.reshape(b, (int(numpy.sqrt(b.shape[0])),int(numpy.sqrt(b.shape[0]))))
    return distance_riemann(a,b)

def build_xy(covs, dist_death):
    x = []
    y = []
    for i in range(len(dist_death)):
        x.append(covs[i].flatten())
        dist = dist_death[i]
            
        j = 0
        f = False
        for la, lb in neighbors_classes:
            if la <= dist and dist < lb:
                y.append(j)
                f = True
                break
            j += 1
        if f is False:
            y.append(len(neighbors_classes)-1)
    return x, y


def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, numpy.newaxis]


    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = numpy.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

matchs = pandas.read_csv('mimic3-scripts-master/result.csv', delimiter=';', header=None, names=['record_id', 'subject_id', 'hadm_id', 'admittime', 'dischtime', 'deathtime', 'hospital_expire_flag', 'accp', 'angus', 'sepsis3', 'start_datetime', 'length'])
# print(matchs)
riemann_dir = "mimic3wdb_riemannv3"
my_k = 7
for f_sigs in sorted(os.listdir(riemann_dir), key=lambda e: 1/len(e)):
    sigs = f_sigs.split('-')
    for f_length in sorted(os.listdir(riemann_dir + '/' + f_sigs), key=lambda e: 1/int(e[:-5])):
        length = f_length[:-5]
        if int(length) < 60*60*18:
            continue
        print(riemann_dir + '/' + f_sigs + '/' + f_length)
        contents = json.load(open(riemann_dir + '/' + f_sigs + '/' + f_length))
        
        covss = []
        for signames, values in contents.items():
#             try:
            values_clean = [v for v in values if (not numpy.isnan(v[3]).any()) and numpy.all(numpy.linalg.eigvals(v[3]) > 0)]

            dist_death = []
            k = False
            for p0x, p0xxxxx, record_id, cov in values_clean:
                tmp = matchs[matchs.record_id == record_id]
                if tmp.shape[0] > 0:
                    if (tmp.hospital_expire_flag == 0).any():
                        dist_death.append(365*24*60*60)
                    else:
                        a, b = tmp.start_datetime.values[0], tmp.deathtime.values[0]
                        dist_death.append(b-a)
                    k = True
                else:
                    dist_death.append(-1)
            if k is False:
                continue
#                 print(dist_death)
#                 dist_death = numpy.ma.masked_where(numpy.isnan(dist_death), dist_death)

            covs_a = numpy.asarray([v[3] for v in values_clean])
            covs_b = numpy.asarray([numpy.asarray(v[3])[2:,2:] for v in values_clean])
        
            if covs_a.shape[0] < 2:
                continue
            
#             pair_dist_a = pairwise_distance(covs_a, metric='riemann') # Ntrials x Nchannels X Nsamples
#             pair_dist_b = pairwise_distance(covs_b, metric='riemann') # Ntrials x Nchannels X Nsamples
            
#             tsne = TSNE(metric='precomputed', perplexity=50, learning_rate=200, early_exaggeration=4, random_state=666)
#             out_a = tsne.fit_transform(pair_dist_a)
#             out_b = tsne.fit_transform(pair_dist_b)

#             plt.figure(figsize=(15,5))
#             plt.subplot(121)
#             plt.scatter(out_a[:, 0], out_a[:, 1], c=dist_death, cmap=palette, norm=norm, s=7)
#             plt.title('[{}+HR+Rv] {} with length {}'.format(f_sigs, signames, beautify_length(int(length), 125)))

#             plt.subplot(122)
#             plt.scatter(out_b[:, 0], out_b[:, 1], c=dist_death, cmap=palette, norm=norm, s=7)
#             plt.title('[{}] {} with length {}'.format(f_sigs, signames, beautify_length(int(length), 125)))
#             plt.colorbar()
#             plt.show()
            
            if (covs_b.shape[0] > 5):
                x, y = build_xy(covs_a, dist_death)
                
                counts = numpy.unique(y, return_counts=True)
                if len(counts[0].tolist()) < 2 or min(counts[1].tolist()) < my_k:
                    continue
#                 print('Examples/class:')
#                 print('  - ', counts[0].tolist())
#                 print('  - ', counts[1].tolist())
                
                plt.figure(figsize=(15,5))
        
                nbrs = KNeighborsClassifier(n_neighbors=my_k, metric=my_dist)
                nbrs.fit(x, y)
                z = nbrs.predict(x)
                counts = numpy.unique(y, return_counts=True)
                cnf_matrix = confusion_matrix(y, z)
                classes_names = numpy.asarray(neighbors_classes_names)[numpy.unique(numpy.concatenate([y,z]))]
                plt.subplot(121)
                plot_confusion_matrix(cnf_matrix, classes=classes_names, normalize=True, title='[{}+HR+Rv] {} with length {}'.format(f_sigs, signames, beautify_length(int(length), 125)))
                report1 = classification_report(y, z, target_names=classes_names)

                
                cnf_matrix = confusion_matrix(y, z)
                x, y = build_xy(covs_b, dist_death)
                nbrs = KNeighborsClassifier(n_neighbors=my_k, metric=my_dist)
                nbrs.fit(x, y)
                z = nbrs.predict(x)
                cnf_matrix = confusion_matrix(y, z)
                classes_names = numpy.asarray(neighbors_classes_names)[numpy.unique(numpy.concatenate([y,z]))]
                plt.subplot(122)
                plot_confusion_matrix(cnf_matrix, classes=classes_names, normalize=True, title='[{}] {} with length {}'.format(f_sigs, signames, beautify_length(int(length), 125)))
                report2 = classification_report(y, z, target_names=classes_names)
                
                print(report1)
                print(report2)
                plt.show()
            
            
#             except Exception as e:
#                 print(e)
#                 pass
                
#                 plt.scatter(out[:, 0], out[:, 1])
#                 plt.title('[{}(+HR+Rvalues)] {} with length {}'.format(f_sigs, signames, length))
#                 plt.show()
```

```python
riemann_dir = "mimic3wdb_riemannv3"

todo = {}

for f_sigs in os.listdir(riemann_dir):
    sigs = f_sigs.split('-')
    for f_length in sorted(os.listdir(riemann_dir + '/' + f_sigs)):
        length = f_length[:-5]
        print(riemann_dir + '/' + f_sigs + '/' + f_length)
        contents = json.load(open(riemann_dir + '/' + f_sigs + '/' + f_length))
        for signames, values in contents.items():
            values_clean = [v for v in values if not numpy.isnan(v[3]).any()]
            for p0x, p0xxxxx, record_id, _ in values_clean:
                if not p0x in todo:
                    todo[p0x] = {}
                if not p0xxxxx in todo[p0x]:
                    todo[p0x][p0xxxxx] = []
                todo[p0x][p0xxxxx].append(record_id)
                

with open('todo_riemannv3.json', 'w+') as f:
    f.write(json.dumps(todo))
```
